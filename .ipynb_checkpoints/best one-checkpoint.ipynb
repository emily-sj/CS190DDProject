# Part 1a

```python
import numpy as np
```

```python
data = np.load("part1a.npz")
```

```python
from scipy import integrate
from itertools import product
import matplotlib as mpl
```

```python
I_0=data['Ic_0']
S_0=data['Svc_0_pmf'].flatten()*(data['N']-sum(I_0))
R_0=np.array([0])##initall recovery is zero 
Lc=data['Lc']
print(Lc.shape)
t=np.linspace(0,120,121) #120 days 
N=data['N'].reshape(-1,1)

Betas=np.linspace(0.1,1.0,16).reshape(4,4)
Gamma=data['gamma'].reshape(-1,1)
initial_vec=np.concatenate([S_0,I_0,R_0,])

```

```python
#build sir model 
def SIR(t,y,betas,gammas):
    S_vec=y[:16].reshape(4,4)
    I_vec=y[16:20]
    R_vec=y[20]
    N=100000 ##population size 
    
    dsvc_dt=-(betas*S_vec*sum(I_vec)/N)
    dic_dt=-(np.sum(dsvc_dt,axis=0))-(gammas*I_vec)
    dr_dt=gammas*sum(I_vec)
    
    vectors=np.concatenate([dsvc_dt.flatten(),dic_dt.flatten(),dr_dt.flatten()])
    return vectors
```

# Part a.1

```python
from scipy.integrate import solve_ivp

solution=solve_ivp(SIR,(t[0],t[-1]),initial_vec,args=(Betas,Gamma),dense_output=True)
solution.sol(t[-1])
```

```python
solutions=[solution.sol(time) for time in t]
S_vc=np.array([np.sum(sol[:16]) for sol in solutions]) 
I_c=np.array([np.sum(sol[16:20]) for sol in solutions])
R=np.array([np.sum(sol[20]) for sol in solutions])

d=I_c.reshape(-1,1)
d1=Lc.reshape(-1,1)
L_over=np.dot(d1,d.transpose())
print(L_over.shape)
virus_overload=np.sum(L_over,axis=0)
```

# Part a.3 Free response 

In our SIR model graph we can depict three different plots. The red plot describes the susceptible, the blue describes the infected, and the green describes the recovered. The blue plot is considered the curve that many people refer as "flatten the curve ". This is because the blue plot depends on the infection rate and if we take precaution we can help flatten the curve, meaning lowering the rate that people get infected. If the blue plot has a high peak, then it will signify a high infection rate at the given time. We could flatten the curve by taking precaution of the virus which will lower the peak and will flatten out to an even area, not presenting a peak at a given time and resulting to less people getting infected as well.

```python
plt.plot(t, S_vc, 'r',label="S plot") # red 
plt.plot(t, I_c, 'b',label="I plot") # blue
plt.plot(t, R, 'g',label= "R plot") # green
plt.legend()
plt.show()


#part a.5 plot sv survivors percentage
end=S_vc[-1]/N.flatten()
print("Not infected population percentage: {}".format(end))
```

We can see that our SIR model plot graph did "flatten" once we changed our betas values by 1/4. The infected plot was never higher than the recovered plot so we assume that the recovery rate was faster than the infection rate which also explains why the susceptible plot never reached to be converged to a number close to zero like our last SIR model. Our original SIR plot graph in which the  infection plot had a high peak lead to a steep negative slope for suspectiable, converging to a low number close to zero. We could assume the reason for the drastic difference in our SIR model given above is that people took caution measures toward the virus which slowed the infection spreading rate which lead to a flatter curve for the infection plot model shown above.

```python
N=50000
Ic_01=data['Ic_0']
Ic_0=np.sum(data['Ic_0'],axis=1)
not_inf=50000-Ic_0 #origin population not infected 
nd1=not_inf[0:5].reshape(-1,1)#trying to format it into corresponding leaf nodes pmf 
nd2=not_inf[5:10].reshape(-1,1)
nd3=not_inf[10:15].reshape(-1,1)
nd4=not_inf[15:20].reshape(-1,1)
#print(nd2)
#print(leaf1.shape)
#print(leaf1)


leaf_nodes=np.array([[.5,.3,.1,.1],[.4,.3,.2,.1],[.3,.3,.2,.2],[.1,.2,.3,.4]])
leafs1=leaf_nodes[0].reshape(-1,1)#corresponding each lead manually 
leafs2=leaf_nodes[1].reshape(-1,1)
leafs3=leaf_nodes[2].reshape(-1,1)
leafs4=leaf_nodes[3].reshape(-1,1)


leaf1=np.dot(nd1,leafs1.transpose())# finding the intial suspectiable population for the given lead nodes 
leaf2=np.dot(nd2,leafs2.transpose())
leaf3=np.dot(nd3,leafs3.transpose())
leaf4=np.dot(nd4,leafs4.transpose())

leaf_node=np.array([leaf1,leaf2,leaf3,leaf4]).reshape(20,4)#all leaf nodes founded 

#print(Ic_01.shape)



L_val=data["L_validation"]
Lc=data['Lc']


betas_v=data["betas_validation"].reshape(-1,1)

Rc=np.array([0])
t=np.linspace(0,19,20)#time interval 
Gamma=data["gamma"]


#initial_vec=np.concatenate([leaf_node.flatten(),Ic_01.flatten(),Rc.flatten()])
#print(initial_vec.shape)
```

```python
def SIR(t,y,betas,gammas):
    S_vec=y[0:4]
    I_vec=y[4:8]
    R_vec=y[8]
    
    N=50000 ##population size 
    
    dsvc_dt=-(betas*S_vec*sum(I_vec)/N)
    dic_dt=-(np.sum(dsvc_dt,axis=0))-(gammas*I_vec)
    dr_dt=gammas*sum(I_vec)
    
    vectors=np.concatenate([dsvc_dt.flatten(),dic_dt.flatten(),dr_dt.flatten()])
    #breakpoint()
    return vectors
```

```python
from sklearn.metrics import mean_squared_error
from scipy.integrate import solve_ivp
beta_grid=np.linspace(0.01,1.0,21)
lf_nodes=20
observed_data=L_val
mmse=np.inf
beta_mmse=np.nan
Rc=np.zeros((lf_nodes))
d=100000
print(beta_grid)
for (beta) in beta_grid:
    #breakpoint()
    cumulative_mse = 0.0
    
    for node_index in range(lf_nodes):
        initial_conditions=np.concatenate([leaf_node[node_index].flatten(), Ic_01[node_index].flatten(), Rc[node_index].flatten()])

        soln_obj=solve_ivp(SIR, (t[0], t[-1]), initial_conditions,\
                                     t_eval=t, args=(beta, Gamma))


        S_predicted=soln_obj.y[0:4]
        I_predicted=soln_obj.y[4:8]
        R_predicted=soln_obj.y[8]
        
        
        L_predicted=np.dot(Lc,I_predicted)
        
        #breakpoint()
        cumulative_mse += mean_squared_error(observed_data[node_index],L_predicted) 
        #breakpoint()
    
    #breakpoint()
    if cumulative_mse<mmse  :
        mmse=cumulative_mse
        beta_mmse=beta
       
        

print(f"Estimated values are beta={beta_mmse:.2f}")
print(mmse)
```
